```python
import torch, json

from LAB3_1.utils import show_loss, show_result
```

```python
results = {}
```

# Assignment 1.1: NARMA10 task with TDNN

```python
fields = torch.load("caches/TDNN_narma10.pt")

results["TDNN_n10"] = dict()
results["TDNN_n10"]["model"] = fields[0]
results["TDNN_n10"]["tr_loss"] = fields[1]

results["TDNN_n10"]["tr_pred"] = fields[2]
results["TDNN_n10"]["tr_y"] = fields[3]
results["TDNN_n10"]["ts_pred"] = fields[4]
results["TDNN_n10"]["ts_y"] = fields[5]

results["TDNN_n10"]["metrics"] = json.load(open("caches/metric_TDNN_narma10.json","r"))
```

## Hyperparameters

1) Hidden units
2) Number of epochs
3) Learning rate
4) Back-step

| Hyperparameters 	|          Values         	|    Description   	|
|:---------------:	|:-----------------------:	|:----------------:	|
|      units      	| 5, 10, 20, 50, 100, 500 	|   Hidden units   	|
|      epochs     	|     10, 20, 50, 100     	| Number of epochs 	|
|        lr       	|  5e-4, 1e-3, 3e-3, 5e-3 	|   Learning rate  	|
|      b_step     	|   5, 8, 10, 12, 15, 20  	|     Back-step    	|


## Model

```python
results["TDNN_n10"]["model"]
```

## Best model & Training loss

```python
print(results["TDNN_n10"]["metrics"]["best_config"])
show_loss(results["TDNN_n10"]["tr_loss"])
```

## Metrics and result

```python
metrics = results["TDNN_n10"]["metrics"]["Model_evaluation"]
print("Model selection")
print(f"Training {metrics[0]} Validation {metrics[1]} Test {metrics[2]}")
print("")
metrics = results["TDNN_n10"]["metrics"]["Final_retrain"]
print("Final retrain")
print(f"Training {metrics[0]} Test {metrics[1]}")

show_result(results["TDNN_n10"]["tr_pred"], results["TDNN_n10"]["tr_y"],
            results["TDNN_n10"]["ts_pred"], results["TDNN_n10"]["ts_y"])
```
